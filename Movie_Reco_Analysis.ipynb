```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
movies.sample(5)
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.sample(5)
```

```python
#movies['overview'] = movies['overview'].apply(lambda x:x.split())

movies.sample(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.sample(5)
```

```python
movies['overview']
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.sample(5)
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

## Now ready for the Model Building

#### In model building we have create a vectorization for the 5000 movies. It means our data convert "TABULER FORMATE" to "VECTOR FORMATE". like this [1,0,0,1..........1]

1. Our data preview is "NEW".
2. Now convert our data from tabluer to vector formate. This process called is Vectorization. 

```python
new
```

#### Import Vectorization libary "CountVectorizer"

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

### Vector shape [1,1,1.......1]

```python
vector.shape
```

### After vectorization we have find out the similerity for similer vector.
for example:  V1:[1,1,0,1,0......1] & V2:[0,1,1,0.....0]  


#### What's the similerity of vector V1 & V2?

#### Now we have use the cosine method.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

### Similarity

```python
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

### Create Recommender function

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Hamlet')
```
